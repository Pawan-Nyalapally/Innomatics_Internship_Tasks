```python
#Task 3: Python Programming Assignment Data Processing and Analysis 
#Problem Statement 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees with highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

```python
#Problem Statement 2: Search Query Keyword Analysis
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation (if any)
import string
for char in string.punctuation:
    query = query.replace(char, "")

# Split into words
words = query.split()

# Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Display words searched more than once
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

```python
#Problem Statement 3: Sensor Data Validation
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid even readings as (hour_index, reading_value)
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem Statement 4: Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Extract domain and count
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage usage
total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Problem Statement 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average_sales = sum(sales) / len(sales)

# Threshold = 30% above average
threshold = average_sales * 1.30

# Detect spike days
for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

```python
#Problem Statement 6: Duplicate User ID Detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

# Count occurrences
for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

# Display duplicates
for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")

```
