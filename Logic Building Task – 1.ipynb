```python
# 2. Pass / Fail Analyzer
marks = [45, 78, 90, 33, 60]

pass_students =0 # count of pass students 
fail_students=0 # count of failed students
for i in marks : 
    if i >=50 :
       pass_students=pass_students +1
    else : 
        fail_students=fail_students +1

print ("total number of pass students : ",pass_students)
print ("total number of fail students :" ,fail_students)
```

```python
#3. Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names :
    cleaned_name = name.strip().lower()  # this Removes spaces & converts to lowercase
    cleaned_names.append(cleaned_name)   # clean name is appended to cleaned_names list 
print(cleaned_names)
```

```python
#4. Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"] # this is a list of messages

for message in messages : 
    length = len(message)
    print (message,"-",length)
    if length >10 :
        print ("This message is longer than 10 characters")
```

```python
#5. Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = logs.count("ERROR")# counts the no of error messages 

print("Total ERROR entries:", error_count) # prints the error count 
```
